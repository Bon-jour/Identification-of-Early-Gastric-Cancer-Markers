```r
data = readRDS('df_all_combined.egc.gc2.rds')
```

```r
data$Group = ifelse(data$Group == 'EGC',1,0)
```

```r
data$Group = factor(data$Group,levels = c(0,1))
```

```r
head(data,2)
```

```r
table(data$physiology)
```

```r
library(autoReg)
overall.log =  glm(Group~. ,data=data,family=binomial())
model3<-autoReg(overall.log,uni=TRUE,milti=T,threshold = 1 , ci = TRUE,  conf.level = 0.95
                 )
```

```r
write.csv(model3,file = 'Grade.uni.csv')
```

```r
model_summary <- summary(overall.log)
coefficient_pvalues <- coef(model_summary)[, "Pr(>|z|)"]
fdr_adjusted_pvalues <- p.adjust(coefficient_pvalues, method = "BH")

# 
corrected_results <- data.frame(
  Variable = names(coefficient_pvalues),
  Odds_Ratio = exp(coef(overall.log)),
  CI_Lower = exp(confint(overall.log)[, 1]),
  CI_Upper = exp(confint(overall.log)[, 2]),
  P_Value = coefficient_pvalues,
  FDR_Adjusted_P = fdr_adjusted_pvalues,
  Significant = fdr_adjusted_pvalues < 0.05
)
```

```r
write.csv( corrected_results,file = 'table2.supplement-S1.csv') 
```
